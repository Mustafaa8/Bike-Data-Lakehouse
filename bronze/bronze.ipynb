```python
import os 
import logging
```

## Load Files to the bronze schema

```python
erp_files = os.listdir("/Volumes/db_ws/bronze/raw/erp")
crm_files = os.listdir("/Volumes/db_ws/bronze/raw/crm")
```

```python
for erp in erp_files:
    erp_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/db_ws/bronze/raw/erp/" + erp)
    erp_df.write.format("delta").mode("overwrite").saveAsTable(f"db_ws.bronze.erp_{erp.split("_")[0]}")

for crm in crm_files:
    crm_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/db_ws/bronze/raw/crm/" + crm)
    erp_df.write.format("delta").mode("overwrite").saveAsTable(f"db_ws.bronze.crm_{crm.split("_")[0]}")
```
