```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
sales = spark.read.table("db_ws.bronze.crm_sales")
sales.display()
```

```python
sales = sales.withColumns({"sls_order_dt":to_timestamp(col("sls_order_dt")),"sls_ship_dt":to_timestamp(col("sls_ship_dt")),"sls_due_dt":to_timestamp(col("sls_due_dt"))})
display(sales)
```

```python
sales.write.mode("overwrite").format("delta").partitionBy("sls_order_dt").saveAsTable("silver_crm_sales")
```
