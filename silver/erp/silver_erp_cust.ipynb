```python
from pyspark.sql.functions import *
```

```python
customers = spark.read.table("db_ws.bronze.erp_cust")
display(customers)
```

```python
customers = customers.withColumn("GEN", when(col("GEN").startswith("M"), lit("Male")).when(col("GEN").startswith("F"), lit("Female")).otherwise(lit("Uknown"))).withColumn("CID", col("CID").substr(-5, 5))
display(customers)
```

```python
customers = customers.withColumnsRenamed({"CID":"customer_id","BDATE":"birthdate","GEN":"gender"})
display(customers)
```

```python
customers.write.mode("overwrite").format("delta").saveAsTable("db_ws.silver.erp_cust")
```
