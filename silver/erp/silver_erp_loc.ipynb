```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
location = spark.read.table("db_ws.bronze.erp_loc")
display(location)
```

```python
location = location.withColumn("CID", col("CID").substr(-5, 5)).withColumnRenamed("CID", "customer_id").withColumnRenamed("CNTRY", "country")
display(location)
```

```python
location.write.format("delta").mode("overwrite").saveAsTable("db_ws.silver.erp_loc")
```
